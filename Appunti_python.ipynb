# 2. **Le variabili**

Iniziamo a conoscere Python, comprendendo un principio fondamentale del linguaggio: le variabili.

Come in altri linguaggi di programmazione, le variabili in Python sono delle locazioni di memoria, dei ‘’contenitori’’ in cui vengono salvati i dati.

Una variabile ha due elementi principali:
- Il **nome**;
- Il **valore** che contiene.

````python
# Per inizializzare una variabile scriveremo
nameAndSurname = "Nicola Milella"
Name = "Mario"
name = "Giacomo"

print(nameAndSurname)
print(Name)
print(name)
````

Per inizializzare correttamente una variabile, ci sono alcune regole da rispettare:
- il nome della variabile può iniziare **solo** con una lettera;
- può contenere **solo** caratteri alfanumerici o underscore;
- **non può** contenere spazi vuoti.

Il comando nella sua interezza è definito **statement**, "dichiarazione", proprio perche' noi siamo andando a dichiarare al nostro linguaggio come deve comportarsi.

Dobbiamo ricordare che Python è un linguaggio a **tipizzazione dinamica**, ovvero il tipo di dato di una variabile non e' definito al momento della dichiarazione ma viene bensi' assegnato automaticamente in base al valore assegnato alla variabile.

Ma soprattutto il tipo di dato puo' cambiare mano a mano che il mio script procede:

````python
# Nella cella superiore, abbiamo detto che name ha come valore una stringa
name = 33

print(name)
````

Importante anche ricordare che Python e' un linguaggio di programmazione **case sensitive**, ovvero differenzia minuscole con maiuscole.

Proprio per questo, come *best practice* si tende ad inizializzare tutte le variabili con iniziale minuscola.

````python
# Creo una variabile Name; mi aspetto come output name = 33 e Name = Roberto
Name = "Roberto"

print(name)
print(Name)
````

## **3.1 Dati semplici**
Cominciamo dai dati semplici: integer, float, string, boolean e complex.
- Per **integer** si intende un numero intero. Viene salvato in una variabile in questo modo:  `num = 10`;
- Per **float**  si intende un numero razionale, a virgola mobile. Viene salvato in una variabile in questo modo: `float = 10.5`;
- Per **string** si intende una sequenza di caratteri con un ordine prestabilito. Viene salvata in questo modo: `string = 'Hello world'`;
- Per **boolean** si intende un dato che può assumere soltanto due valori, *true* e *false*. Viene salvato in questo modo: **bool = True**;
- Per **complex** si intende un numero complesso, ovvero un numero che comprende una parte reale e una immaginaria. Viene salvato in questo modo `complex = 3j`.

Per aiutarci a riconoscere il tipo di dato contenuto all’interno di una variabile, possiamo utilizzare la funzione `type()`:

#### **3.2.2.2 Unpacking dei valori**
Una delle caratteristiche che rende le tuple molto apprezzate è il cosiddetto **unpacking dei valori**, ovvero la possibilità di salvare automaticamente all’interno di alcune variabili ogni singolo elemento della tupla:

````python
punto_3D = (1.34, 4.34, 66.77)

x, y, z = punto_3D

print(f"punto 3D X: {x}, Y: {y}, Z: {z}")

# altro esempio

students_tuple = ( "Giuseppe", "Linda", 43 )

n1, n2, n3 = students_tuple

print(n1)
print(n2)
print(n3)
````

````python
punto_3D = (1.34, 4.34, 66.77)

x, y, z = punto_3D

print(f"punto 3D X: {x}, Y: {y}, Z: {z}")

punto_3D_2 = punto_3D

print(f"punto 3D X: {x}, Y: {y}, Z: {z}")

x, y, z = punto_3D_2

print(f"punto 3D 2 X: {x}, Y: {y}, Z: {z}")
````

````python
# e se avessimo un mismatch?

students_tuple = ( "Giuseppe", "Linda", "Roberta", "Nicola" )

n1, n2, n3 = students_tuple
print(n1)
print(n2)
print(n3)
````

Un'altra azione molto semplice da eseguire con una tupla e' un algoritmo noto come swap di due variabili:

````python
# senza tuplas
x = 5
y = 10

print(f"Prima dello swap, x = {x} e y = {y}")

temp = x
x = y
y = temp

print(f"Dopo lo swap, x = {x} e y = {y}")
````

````python
#con tupla
x = 5
y = 10

print(f"Prima dello swap, x = {x} e y = {y}")

x, y = (y, x)

print(f"Dopo lo swap, x = {x} e y = {y}")
````

### **3.2.4 Set**
Un'altra struttura dati è il **Set**, un tipo di dato che ha tantissimo in comune con gli insiemi matematici.

La dichiarazione di un set può sembrare simile a quella dei dizionari ma in realtà è profondamente diversa: è vero che si utilizzano lo stesso le parentesi graffe, ma non avremo la classica struttura chiave-valore.

````python
animals = {'dog', 'cat', 'wolf'}

# oppure

animals = set(['dog', 'cat', 'wolf'])
````

Ma perché abbiamo specificato che il set ha molto in comune con gli insiemi matematici?

Perché un set in Python è un insieme di elementi univoci, non ordinati e immutabili:

````python
animals = {'dog', 'cat', 'wolf', 'dog'}

print(animals)
````

#### **3.2.4.1 Metodi dei set**
Vediamo alcuni metodi specifici dei set 

````python
s = {'Nicola', 37, 1.75}

print(s)
````

````python
#NB: l'ordinamento è dato dall'hash delle chiavi del set
````

````python
# set - trasforma una lista in un set
l = ['dog', 'wolf', 'cat', 'cat', 'dog', 'bear']
set_from_list = set(l)
print(set_from_list)

# add - aggiunge un elemento al set
s.add("Python")
print(s)

# remove - rimuove un elemento dal set
s.remove(1.75)
print(s)

# intersection - ottiene i valori comuni tra due set
s = {"Nirvana", "Blind Guardian", "Litfiba", "Subsonica"}
s2 = {"Daugher", "London Grammar", "Subsonica", "Nirvana"}
s3 = s & s2
print("intersezione", s3)

# difference - ottiene i valori presenti in un set e non nell'altro
s = {"Nirvana", "Blind Guardian", "Litfiba", "Subsonica"}
s2 = {"Daugher", "London Grammar", "Subsonica", "Nirvana"}
print("differenza", s2.difference(s))

# union - ottiene i valori di tutti i set
s = {"Nirvana", "Blind Guardian", "Litfiba", "Subsonica"}
s2 = {"Daugher", "London Grammar", "Subsonica", "Nirvana"}
s3 = s.union(s2)
print("union", s3)
s4 = {"Daugher", "London Grammar", "Subsonica", "Nirvana", "a","b"}
#issubset - controlla se i valori di un set sono presenti all'interno di un altro set
s2.issubset(s4)
````

Con gli **operatori aritemici** possiamo fare dei semplici calcoli matematici:

## **4.2 Operatori di confronto**

````python
table = PrettyTable()
table.add_column("Simbolo", ["<", ">", "<=", ">=", "==", "!="])
table.add_column("Nome operazione", ["Minore", "Maggiore", "Minore o uguale", "Maggiore o uguale", "Uguale", "Diverso"])


print(table)
````

## **4.3 Operatori di assegnazione**

Parliamo adesso di uno dei fondamentali di ogni linguaggio di programmazione. Sappiamo che un algoritmo viene letto dall’inizio alla fine ma, alle volte, abbiamo la necessità che l’algoritmo si comporti diversamente in base al valore di una condizione.

Cominciamo dunque a parlare di **selezione**.

Per selezione si intende, appunto, la scelta fra due percorsi da seguire successivamente, che dipende da una condizione che può essere vera o falsa.

Si ottiene principalmente con lo statement **if**, dove verrà eseguito un set di istruzioni laddove la condizione sarà vera. In caso di condizione falsa invece, il codice non eseguira' alcun codice, andando avanti.

# **6. L'iterazione (*for*, *while*)**
Un altro dei fondamentali della programmazione è il concetto di **iterazione**, o **ciclo**, ovvero un blocco di istruzioni che vengono ripetutamente eseguite fino a che una certa condizione non cambia di stato.

Python utilizza due tipi di ciclo: il ciclo **for** e il ciclo **while**.

## **6.1 Ciclo For**
Il ciclo **for** in Python e' un costrutto iterativo che consente di eseguire un blocco di codice un numero **determinato** di volte oppure su una sequenza di elementi.

La sintassi è molto semplice. Mettiamo caso volessimo stampare un numero da 0 a 5:

````python
for i in range(5):
#print("fina") # Errore!!!
    print(i)



print('fine ciclo')
````

```javascript
for (let i = 0; i < 6; i++) {
    console.log(i);
}
```

```php
for ($i = 0; $i < 6; $i++) {
    echo $i;
}
```

````python
l = [10,20,30,40,50];

for e in l:
  print(e)


string = "Hello world!"

for char in string:
  print(char)

print('fine ciclo')
````

Quindi un *for* statement e' formato da:
- `for`: indica l'inizio del ciclo;
- `i`: e' la variabile di iterazione che assume i valori successivi della sequenza;
- `in`: segna l'inizio della sequenza su cui iterare;
- `range(1, 6)`: nel nostro caso, la sequenza su cui iterare;

Dobbiamo anche ricordare che la funzione `range()` non include mai il numero passato in posizione di end.

La funzione `range()` accetta pero' tre argomenti, di cui due sono impliciti: noi abbiamo passato la fine del range ma possiamo passare anche l'inizio e il valore dell'incremento.

Volendo per esempio stampare i numeri da 0 a 6, ma incrementando l’indice di due scriveremo:

````python
for i in range (2, 17, 3):
  print(i)

print('fine ciclo')
````

E se volessimo includere gli estremi?

Proviamo a stampare tutti i numeri da -2 a 2:

````python
min = -2
max = 2

for i in range(min, max + 1):
  print(i)

print('fine ciclo')
````

In questo caso, l’espressione matematica viene eseguita prima di creare il range.

Non siamo però limitati alla funzione `range()` quando utilizziamo il *for*.

Infatti, dopo la keyword in possiamo inserire qualsiasi tipo di **dato iterabile**, tra cui *string*, *list*, *tuple*, *dict* e *set*.   

## **6.2 Break**
Abbiamo detto che il ciclo *for* viene utilizzato quando sappiamo esattamente quante iterazioni vogliamo far effettuare al nostro algoritmo oppure fino alla fine di una certa sequenza di elementi.

In alcuni casi vorremmo poter mettere fine al ciclo al sopraggiungere di determinate condizioni.

Mettiamo caso di voler ciclare una parola fino a trovare una lettera maiuscola:

````python
password = "passWord"

for letter in password:
  print(letter)

  if letter.isupper():
    print(f"Una maiuscola! Fermiamo il ciclo!")
    break

print("FINE")
````

Possiamo notare come il ciclo continui fino a quando, trovata una lettera maiuscola, chiediamo al processo di interrompersi tramite lo statement **break**.

Di solito il *break* viene utilizzato per evitare codice non necessario.

## **6.3 Ciclo While**
Un altro ciclo utilizzato da Python è il while.

Cosa lo distingue dal ciclo for?

Che, invece di far riferimento a una quantità precisa di iterazioni, il while si basa su una espressione booleana, quindi su una condizione: se questa condizione sarà True, comincerà o continuerà il ciclo.

Mettiamo caso io volessi stampare dei numeri da 1 a 5:

````python
i = 1

while i <= 5:
    print(i)
    i += 1

print("fine ciclo")
````

````python
n = int(input('inserisci un numero'))

while n != 1 :
    print(n)
    if(n % 2 == 0):
        n = n // 2
    else:
        n = n * 3 + 1


print(f"Ciclo terminato N e' {n}")
````

Al contrario del for, possiamo notare che contatore e incrementatore sono impliciti.

# **9. Exception Handling**
Una costante di ogni algoritmo e di ogni linguaggio di programmazione sono gli errori: è virtualmente impossibile riuscire a scrivere un codice complesso funzionante, senza alcun tipo di errore, al primo colpo.

Adesso, ovviamente, ci concentreremo su tutti quegli errori che è possibile trovare in un programma Python.


Gli errori su Python, generalmente, appartengono a due grandi categorie:

- **errori di sintassi**, ovvero quando scriviamo in maniera non corretta delle istruzioni Python. Importante è ricordare che, quando avviene un errore di sintassi, Python blocca totalmente l’esecuzione dell’algoritmo.

````python
a = 6

if a > 4
    print('è maggiore')

# Abbiamo scordato di mettere i : dopo la condizione
````

- **errori a runtime**, quando l’errore accade durante l’esecuzione del programma. Un errore a runtime ci indica che c’è un flusso logico che il nostro programma non ha considerato e ci porta quindi a parlare di eccezione.

````python
a = 4
b = 0

print("CIoa")
print(a / b)

print("non continua")
# Stiamo provando a dividere un numero per 0
````

Immaginiamo che `a` e `b` siano frutto dell’input da parte di un utente e il mancato controllo su questi valori potrebbe causare un errore a runtime, ovvero una possibilità che da sviluppatori potevamo prevedere ma che non abbiamo previsto.

Vuol dire che noi possiamo sempre agire/reagire per prevenire e meglio gestire questo tipo di errori.

Ma come facciamo a gestire le eccezioni?

## **9.1 Try Except**
Esiste un costrutto Python chiamato **Try Except**, che ci aiuta appunto a gestire questo tipo di errori:

````python
#try:
    # Prova ad eseguire una serie di istruzioni
#except:
    # Vai a eseguire le righe specificate qui se avviene un errore a runtime
````

Con questo costrutto possiamo gestire l'errore di prima:

````python
a = 4
b = 0

c = a/b
try:
    print("eseguo il blocco")
    print(a / b)
    print("fine blocco")
except:
    print('Non puoi eseguire questa operazione')

print("continua")
````

A cosa serve? Serve per fare in modo che il nostro script funzioni anche se un blocco di codice causa un errore, perché quell’errore è stato gestito da noi sviluppatori.

Proviamo a fare qualcosa di leggermente più complicato, sfruttando un ciclo e l’input di un utente:

````python
while True:
    x = input('inserisci un numero (0 per chiudere): ')
    if int(x) == 0:
        break
    else:
        print('Continua...')
````

Con questo blocco di codice, Python proseguirà col ciclo fino a quando non verrà inserito un numero uguale a 0.

Ma se l’utente inserisse una lettera?

Si bloccherebbe tutto, in quanto la funzione `int()` non può gestire una stringa. Possiamo evitare che l’utente inserisca una stringa? Assolutamente sì:

````python
while True:
    x = input('inserisci un numero (0 per chiudere): ')
    try:
        if int(x) == 0:
            break
        else:
            print('Continua...')
    except:
        print('Hai inserito un valore non numerico. Riprova.')
````

Cosa stiamo facendo? Inserendo la struttura Try Except, abbiamo la possibilità di gestire l’errore in runtime: nel caso l’utente inserisse un valore non numerico, verrà eseguito il blocco di codice in except, in quanto ha catturato l’errore di cui sopra, e questo permetterà al ciclo di continuare senza che il suo flusso venga interrotto.

## **9.2 Built-in Exeptions**
Python al suo interno ha delle eccezioni built-in, ovvero delle casistiche d’errore standard (`ZeroDivisionError` è un esempio; `ValueError` ne è un altro).

In realtà queste eccezioni sono delle vere e proprie classi che noi possiamo catturare specificatamente all’interno di una struttura Try Except:

````python
while True:
    x = input('inserisci un numero (0 per chiudere): ')
    try:
        if int(x) == 0:
            break
        else:
            print('Continua...')
    except ValueError:
        print('Hai inserito un valore non numerico. Riprova.')
````

Aggiungendo alla keyword `except` il nome dell’eccezione, possiamo catturare solo gli errori di quella classe . In questo caso specifico, noi stiamo dicendo a Python che deve stampare la nostra stringa solo nel caso catturasse un errore di classe ValueError.

Ma possiamo gestire in maniera specifica ogni singolo errore a cui possiamo pensare? Assolutamente sì.

Diamo la possibilità all’utente di far scattare un errore di classe `ZeroDivisionError`:

````python
while True:
    x = input('inserisci un numero per effettuare una divisione (scrivi quit per chiudere): ')
    if x == 'quit':
      break;
    try:
        x = int(x)
        print(10 / x)
        if int(x) == 0:
            break
        else:
            print('Continua...')
    except ValueError:
        print('Hai inserito un valore non numerico. Riprova.')
    except ZeroDivisionError:
        print('Non puoi effettuare una divisione per zero')
````

Abbiamo anche la possibilità di gestire un’eccezione generica:

````python
while True:
    x = input('inserisci un numero per effettuare una divisione (scrivi quit per chiudere): ')
    if x == 'quit':
      break;
    try:
        x = int(x)
        print(10 / x)
        if int(x) == 0:
            break
        else:
            print('Continua...')
    except ValueError:
        print('Hai inserito un valore non numerico. Riprova.')
    except ZeroDivisionError:
        print('Non puoi effettuare una divisione per zero')
    except Exception:
        print('Hai inserito un valore non valido')
````

Quindi quello che succede è che se viene incrociato un ValueError sarà lanciato il blocco di codice di quell’except; altrimenti quello di ZeroDivisionError; altrimenti ancora quello generico.

La cosa da ricordare è che gli `except` vengono controllati in sequenza: se mettessimo a testa l’eccezione generica, verrà lanciato sempre quel blocco di codice. Quindi come best practice,
inseriamo sempre in cima le eccezioni più specifiche, mentre come ultimo blocco di codice inseriamo quella generica, in modo tale da avere un controllo più granulare sul nostro codice.
