# **1. Le funzioni**

Parliamo adesso di un altro dei fondamentali di ogni linguaggio di programmazione, ovvero le **funzioni**.

Una **funzione** è un costrutto che permette di raggruppare una sequenza di istruzioni in un unico blocco che può essere richiamato più volte all’interno del nostro algoritmo.

Solitamente restituisce un valore o **output** all’utente.

Esistono due tipi di funzioni:
- le **built-in functions** (BIF), ovvero tutte quelle funzioni che sono inglobate all’interno del linguaggio di programmazione, create dagli sviluppatori dello stesso. La funzione `print()` o `isupper()` ne sono un esempio;
- le **User functions**, ovvero tutte quelle funzioni create all’interno dell’algoritmo dall’utente.

Ma come si crea una funzione?

Il primo passo è la **dichiarazione**. Useremo la keyword **def** seguita dal nome che vorremo dare alla funzione; in seguito andremo a **esplicitare le istruzioni** che compongono il corpo della funzione; se necessario, andremo **esplicitare i parametri** che deve manipolare quando la richiamiamo all’interno dell’algoritmo.

Proviamoci:

```python
def say_hello(): #dichiarazione funzione
    print("Hello world!") #corpo della funzione
    print("Ciao mondo!")
    print("FINE FUNZIONE\n")
```

```python
# Cosi sarebbe il mondo senza funzioni!! :(
print("Hello world!") #corpo della funzione
print("Ciao mondo!")
print("FINE FUNZIONE\n")

print("Hello world!") #corpo della funzione
print("Ciao mondo!")
print("FINE FUNZIONE\n")

print("Hello world!") #corpo della funzione
print("Ciao mondo!")
print("FINE FUNZIONE\n")


print("Hello world!") #corpo della funzione
print("Ciao mondo!")
print("FINE FUNZIONE\n")
```

## **1.1 I parametri**
Abbiamo parlato di parametri, ma cosa sono?

Con **parametro** si intende l’insieme dei dati che la funzione si aspetta di ricevere per eseguire il suo compito. Sono come delle variabili speciali che vengono definite all'interno della funzione e che assumono i valori passati quando la funzione viene chiamata.

Esistono due tipi di parametri:
- **Parametri formali**: Sono i segnaposto all'interno della funzione. Vengono definiti nella riga di definizione della funzione e specificano il tipo e il nome dei dati che la funzione si aspetta di ricevere;
- **Argomenti attuali o reali**: Sono i valori effettivi che vengono passati alla funzione quando viene chiamata. Vengono specificati tra parentesi all'atto della chiamata.

```python
x = 5
y = 10

def sum(a, b): # qui a e b sono i parametri formali
    print(a + b)
   

sum(x, y) # qui x e y sono i parametri attuali di questa chiamata
sum(55, 99) # qui 55 e 99 sono i parametri attuali di questa chiamata
```

```python
def divide(a, b):
  print(a / b)

divide(x, 0)
divide(y, x)
```

## **1.2 Parametro di Default**
Avendo detto che i parametri funzionano tramite la loro posizione, nel caso in cui noi nella funzione dichiarassimo tot parametri formali dovremo necessariamente inserire in seguito pari parametri reali.

Passandone di meno succede questo:

```python
sum(x)
```

Abbiamo però la possibilità di impostare nella dichiarazione della funzione dei **parametri di default**.

Il parametro di default scatterà nel caso in cui non venisse passato il giusto numero di parametri reali.

Nel caso, invece, venisse passato il giusto numero di parametri reali, il parametro di default verrà semplicemente ignorato.

```python
# x = 5
# y = 10

def sum(a, b = 80):
    print(a + b)

sum(1,2)
sum(x) # qui avro' 85!
```

## **1.3 Keyword Arguments**
Finora abbiamo gestito i nostri parametri tramite la loro posizione.

In Python pero' abbiamo un altro modo, piuttosto flessibile, per passare argomenti alle funzioni in Python noto come **Keyword Arguments**.

In questo caso, invece di affidarsi esclusivamente all'ordine degli argomenti, possiamo specificare esplicitamente quale valore corrisponde a ciascun parametro formale della funzione.

```python
a_name = "Giancarlo"
a_surname = "Valente"

def sayHello(surname, name):
  print(f"Ciao, mi chiamo {surname}, {name} {surname}")


sayHello("Bond","James")

sayHello(name = a_name, surname = a_surname)
sayHello(surname = a_surname, name = a_name)
sayHello(surname = "Bond", name = "James")
sayHello( name = "James",surname = "Bond")
```

## **1.4 Parametri variabili**
Ci sono rari casi in cui non sappiamo quanti parametri verranno passati alla nostra funzione.

In questi casi si utilizzerà il cosiddetto **unpack operator**: all’interno dei parametri formali utilizzeremo il carattere `*` precedente al parametro.

Possiamo usarlo in due casistiche:

```python
def sum(a, b, c):
  print(f"La somma e':", a + b + c)

numbers = [1, 2, 3]
sum(*numbers)



# l'operatore di unpack espande la lista numbers in tre argomenti separati, ovvero 1, 2, e 3 che poi vengono usati al posto di a, b e c.
```

```python
x = 5
y = 10
z = 20
w = 30
a = 5

def sum(*numbers):
    print(numbers)
    print(type(numbers))
    result = 0
    
    for number in numbers:
        result += number
        
    print(f"La somma e': {result}")

sum(x, y, z, w,a,34,5665,8)
sum(x, y,a,34,5665,8)
sum(x, y, z, w,5665,8)
# in questo caso usiamo l'operatore di unpack per indicare alla funzione che puo' accettare un numero variabile di argomenti
```

## **1.5 Return statement**
Prendendo ad esempio la funzione precedente, possiamo facilmente comprendere che riusciamo a vedere il risultato solo in quanto nella funzione abbiamo utilizzato `print()`. Togliendo `print()` non riusciremmo piu' a vedere il risultato:



```python
x = 5
y = 10
z = 20

def sum(*numbers):
  result = 0

  for number in numbers:
    print(number)
    result += number

print(sum(x, y, z))
```

Per ottenere il risultato sperato dobbiamo utilizzare lo statement **return**, un costrutto fondamentale utilizzato all'interno delle funzioni per restituire un valore o un oggetto alla parte di codice che ha chiamato la funzione.

```python
x = 5
y = 10
z = 20

def sum(*numbers):
    result = 0

    for number in numbers:
        result += number
        
    return result

print(f"Il risultato della somma e':", sum(x, y, z))

# in questo modo possiamo salvare anche il valore in una variabile

sum = sum(x, y, z)
print(f"Il risultato salvato nella variabile e': {sum}")

# print(f"Il risultato della somma e':", sum(x, y, z)) # Va in errore!!! ATTENZIONE !!!
```

# **2. Mutability in Python**
Parliamo adesso del concetto di **Mutabilità** in Python.

Prima di partire, però, abbiamo bisogno di fare un passo indietro: Python poggia su un’implementazione che si basa su C, chiamata **CPython**.

*CPython* può essere definito contemporaneamente un interprete e un compilatore, in quanto prende il nostro codice sorgente Python (definito **codice di alto livello**) e lo trasforma in *bytecode* (definito **codice di livello intermedio**) salvandolo con estensione `.pyc`.

Una volta trasformato, il codice verrà eseguito tramite la **Python Virtual Machine**.

In *CPython*, ogni tipo di dato ha in realtà una struttura complessa che lo descrive, ed è quella che viene salvata all’interno della memoria, non il valore in sé per sé della variabile.

E’ per questo che ogni tipo di dato, anche il più semplice, è considerato come un oggetto.


La mutabilita' si riferisce alla capacita' di un oggetto di modificare il suo stato dopo la creazione. In altre parole:
- un oggetto mutabile puo' essere modificato;
- un oggetto immutabile rimane lo stesso una volta creato.

Ma da cosa deriva questo meccanismo?

Tutto sta in come Python gestisce il comportamento dei dati in memoria.

Infatti, i tipi mutabili vengono memorizzati come riferimenti ad un'area di memoria che contiene i dati effettivi: la modifica dell'oggetto quindi modifica i dati in memoria a cui si riferisce.

I tipi immutabili, invece, vengono memorizzati direttamente in memoria: la modifica di un valore immutabile crea un nuovo oggetto con il nuovo valore, lasciando inalterato l'oggetto originale.

Immaginiamo di essere in un ristorante e di ordinare un piatto di pasta:
- il piatto di pasta e' immutabile: il suo contenuto non puo' essere modificato una volta che e' stato preparato in cucina e di solito il piatto ha un nome con cui viene identificato (*carbonara*, *bistecca* ecc);
- l'ordine del cameriere e' invece mutabile: infatti il cameriere puo' aggiungere o modificare i piatti che hai ordinato. L'ordine pero' non contiene il cibo stesso, ma solo i **riferimenti** ai piatti che hai ordinato, identificati dal loro nome.





I tipi di dato **mutabili** sono:
- liste;
- dizionari;
- set;
- oggetti generici.

I tipi di dato **immutabili** sono:
- stringhe;
- tuple;
- integer;
- numeri reali;
- None.

Se pensiamo all'esempio del ristorante, possiamo notare che i dati mutabili sono dei "contenitori" (liste, dizionari...); i dati immutabili invece indicano dei dati "effettivi" con un valore "effettivo" (stringhe, numeri...).

Ma come possiamo toccare con mano il comportamento che Python ha nella memoria?

Ci viene in aiuto il metodo `id()`, il cui compito e' accedere proprio all'identificativo in memoria di un oggetto specifico.

```python
# Tipi mutabili - lista
list = [1, 2, 3]

# stampo il contenuto della lista
print(list)

# stampo l'id in memoria della lista
print(id(list))

# stampo l'esadecimale dell'id in memoria della lista
print(hex(id(list)))
```

```python
# Andiamo a modificare la lista
list[1] = 10

# stampo il contenuto della lista
print(list)

# stampo l'id in memoria della lista
print(id(list))

# stampo l'esadecimale dell'id in memoria della lista
print(hex(id(list)))
```

Possiamo notare che, nonostante i valori all'interno della lista siano cambiati, l'indirizzo di memoria a cui e' collegata quella lista e' rimasto invariato.

Cosa accade pero' quando lavoriamo con dati immutabili, come le stringhe?

```python
# Tipi immutabili - stringa
string = "Hello world!"

# stampo il contenuto della stringa
print(string)

# stampo l'id in memoria della stringa
print(id(string))

# stampo l'esadecimale dell'id in memoria della stringa
print(hex(id(string)))
```

```python
# Proviamo a modificare la stringa
string = string + ", Goodbye world?"

# stampo il contenuto della stringa
print(string)

# stampo l'id in memoria della stringa
print(id(string))

# stampo l'esadecimale dell'id in memoria della stringa
print(hex(id(string)))

# Usate Oggetti che si chiamano StringBuilder !!!
```

## **2.1 La mutability nelle funzioni**
Quando andiamo a lavorare con le funzioni, come vengono passati gli argomenti e quali sono le conseguenze per oggetti mutabili e immutabili?

Il modo in cui il compilatore Python gestisce gli argomenti delle funzioni è strettamente correlato al fatto che essi siano mutabili o no.

Normalmente, un oggetto mutabile viene passato **per riferimento**, ovvero la variabile originale può essere modificata (in quanto alla funzione viene dato accesso direttamente alla cella di memoria). Se si desidera evitare questo comportamento, è necessario copiare la variabile in un’altra variabile.

Al contrario, un oggetto immutabile viene passato a una funzione **per valore**, quindi non è possibile andare a modificare intrinsecamente il valore di quell’immutabile (in quanto alla funzione viene passata solo una copia del valore contenuto nella variabile).

```python
def increment(x):
  x += 1

x = 5

print(f"Il valore di 'x' prima dell'incremento e': {x}")
increment(x)
print(f"Il valore di 'x' dopo l'incremento e': {x}")
```

Come sappiamo, un integer in Python è immutabile, quindi la funzione increment() non ha alcun effetto diretto sul valore di x.

Proviamo adesso invece a utilizzare un dato mutabile, passandolo a una funzione che mi faccia delle modifiche:

```python
def edit_name(dict):
  dict['first_name'] = "Roberto"

teacher = {'first_name': 'Giancarlo', 'last_name': 'Valente'}

print(f"Il contenuto di 'teacher' prima della modifica e' {teacher} con id {hex(id(teacher))}")
edit_name(teacher)
print(f"Il contenuto di 'teacher' dopo la modifica e' {teacher} con id {hex(id(teacher))}")
```

Particolare è invece il caso della tupla: abbiamo detto che la tupla è un oggetto immutabile ma questa dichiarazione è vera solo in parte.

La tupla in sé per sé non può essere modificata ma, in alcuni casi, è possibile modificare gli oggetti in essa contenuti. Infatti, se la tupla ha un campo immutabile come una stringa, allora la tupla non può essere modificata (parliamo di **immutabilità non transitiva**); ma un campo mutabile come una lista, anche se contenuto all’interno di una tupla, può comunque essere modificato:

```python
# la tupla e' immutabile

t = ("Davide", "Roberto")
print(f"L'id della prima tupla e' {hex(id(t))}")
print(t)
t = ("Davide", "Roberto")
print(f"L'id della prima tupla e' {hex(id(t))}")

print(t)
```

```python
# se proviamo a modificare un campo immutabile (la stringa) ci dara' errore

t[1] = "Francesco"
```

```python
# ma cosa succede con un campo mutabile dentro la tupla?

t = (1, 2, ['Hello', 'World'])

print(f"L'id della tupla prima della modifica e' {hex(id(t))}")

t[2].append('!!!')

print(f"L'id della tupla dopo la modifica e' {hex(id(t))}")

print(f"Il contenuto della tupla e' {t}")
```
