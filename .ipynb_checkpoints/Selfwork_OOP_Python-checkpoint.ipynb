```python
# Selfwork Python - OOP
# Esercizi:
# 1) Creare una classe Person , estenderla con due sottoclassi Dipendente e Studente
# 2) Estendere Dipendenti con 2 sottoclassi Docente e Freelance.
# Una classe deve essere astratta e ogni classe deve avere delle proprietà specializzanti.

from abc import ABC, abstractmethod


class Person(ABC):
    def __init__(self, nome: str, cognome: str, eta: int):
        self.nome = nome
        self.cognome = cognome
        self.eta = eta

    # Metodo astratto: obbliga le sottoclassi a implementare una descrizione personalizzata
    @abstractmethod
    def get_details(self) -> str:
        pass



class Studente(Person):
    def __init__(self, nome: str, cognome: str, eta: int, matricola: str, corso_studi: str):
        # Ereditiamo i parametri dalla classe madre
        super().__init__(nome, cognome, eta)
        # Proprietà specializzanti
        self.matricola = matricola
        self.corso_studi = corso_studi

    def get_details(self) -> str:
        return f"Studente: {self.nome} {self.cognome} | Matricola: {self.matricola} | Corso: {self.corso_studi}"


# 3. SOTTOCLASSE DI PERSON (Base per i lavoratori)
class Dipendente(Person):
    def __init__(self, nome: str, cognome: str, eta: int, id_azienda: str, stipendio_base: float):
        super().__init__(nome, cognome, eta)
        # Proprietà specializzanti per tutti i dipendenti
        self.id_azienda = id_azienda
        self.stipendio_base = stipendio_base


    def get_details(self) -> str:
        return f"Dipendente {self.id_azienda}: {self.nome} {self.cognome}"


# 4. SOTTOCLASSE DI DIPENDENTE
class Docente(Dipendente):
    def __init__(self, nome: str, cognome: str, eta: int, id_azienda: str, stipendio_base: float, materia: str, dipartimento: str):
        super().__init__(nome, cognome, eta, id_azienda, stipendio_base)
        # Proprietà per il docente
        self.materia = materia
        self.dipartimento = dipartimento

    def get_details(self) -> str:
        return f"Docente di {self.materia} ({self.dipartimento}): {self.nome} {self.cognome} | ID: {self.id_azienda}"


# 5. SOTTOCLASSE DI DIPENDENTE
class Freelance(Dipendente):
    def __init__(self, nome: str, cognome: str, eta: int, id_azienda: str, stipendio_base: float, partita_iva: str, tariffa_oraria: float):
        super().__init__(nome, cognome, eta, id_azienda, stipendio_base)
        # Proprietà per il freelance
        self.partita_iva = partita_iva
        self.tariffa_oraria = tariffa_oraria

    def get_details(self) -> str:
        return f"Freelance P.IVA {self.partita_iva}: {self.nome} {self.cognome} | Tariffa: €{self.tariffa_oraria}/h"


studente1 = Studente("Sofia", "Bianchi", 21, "S12345", "Informatica")
docente1 = Docente("Alessandro", "Verdi", 45, "D9981", 2500.0, "Programmazione Python", "Scienze MM.FF.NN.")
freelance1 = Freelance("Emma", "Neri", 33, "F5542", 0.0, "IT12345678901", 50.0)

print(studente1.get_details())
print(docente1.get_details())
print(freelance1.get_details())
```
